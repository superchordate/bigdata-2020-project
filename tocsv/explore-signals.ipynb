```python
import pandas as pd
import matplotlib.pyplot as plt
import pyedflib # pip install git+https://github.com/superchordate/pyedflib
import random
```

```python
file = 'rbd13'
edfpath = '../edf/'
edf = pyedflib.EdfReader(edfpath + file + '.edf')
info = pd.DataFrame({
    'label': edf.getSignalLabels(),
    'sample_rate': edf.getSampleFrequencies()
})
```

```python
def plotsignal_file(file, channel, startat = 0, step = 10000):
    edf = pyedflib.EdfReader(edfpath + file + '.edf')
    freq_index = edf.getSignalLabels().index(channel)
    signal = edf.readSignal(freq_index)
    plt.figure(figsize=(20, 5))
    plt.plot(signal[startat:(startat + step)])
```

```python
plotsignal_file('rbd13', 'ROC-LOC', 0, 1000000)
```

```python
plotsignal_file('rbd13', 'ROC-LOC', random.randint(0,3000000), 100000)
```
