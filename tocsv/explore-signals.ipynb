```python
def plotsignal(file, startat = 0, step = 10000):
    signals = pd.read_parquet('../pytorch/train/'+ file + '-128.parquet')
    plt.figure(figsize=(20, 10))
    plt.plot(signals['ROC-LOC'][startat:(startat + step)])
```

```python
plotsignal('n2')
```

```python
plotsignal('n2', 500000)
```

```python
plotsignal('rbd13')
```

```python
plotsignal('nfle2')
```

```python
plotsignal('nfle3')
```
